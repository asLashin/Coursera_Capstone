# Segmenting and Clustering Neighborhoods in Toronto .. 

In this notebook, I will explore, segment, and cluster the neighborhoods in the city of Toronto. 

First Step is to import the required packages .. 

```python
import pandas as pd
import numpy as np
```

Create the URL for the Wikipedia page .. 

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

url
```

Using pd.read_html() function to read HTML webpage

```python
df_page = pd.read_html(url, header = 0)
df_page
```

```python
len(df_page)
```

Get the first table that contains the neighborhoods in the city of Toronto


```python
df = df_page[0]

df
```

Remove the cells that don't have an assigned borough. Ignore cells with a borough that is Not assigned.

```python
df['Neighbourhood'].replace('Not assigned', np.nan, inplace = True)

df.dropna(subset = ['Neighbourhood'], axis = 0, inplace = True)

df
```
